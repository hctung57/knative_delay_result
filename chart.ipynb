```python
import handle
import numpy as np
import matplotlib.pyplot as plt
```

```python
DNS_FOLDER = "dns"
NODNS_FOLDER = "nodns"
CUSTOM_FOLDER = "custom"
DEFAULT_FOLDER = "default"
CLOUD_FOLDER = "cloud"
EDGE_FOLDER = "edge"
class DataBlock:
    def __init__(self, mean, error) -> None:
        self.mean = mean
        self.error = error

class DataBlockByDelay:
    def __init__(self, lookup, request, total) -> None:
        self.lookup_block = lookup
        self.request_block = request
        self.total_block = total
        
```

```python
def calculate_mean_and_error_from_list(mesure_location , number_pod, type_dns, type_code, delay, pod_location=None):
    list_time_lookup = []
    list_time_request = []
    list_time_total = []
    
    for i in range(1,11):
        filename = ""
        if number_pod == 1:
            filename = "./{}/{}_{}_{}/{}/{}/data-{}.txt".format(mesure_location,number_pod, type_dns, pod_location,type_code, delay,i)
        else:
            filename = "./{}/{}_{}/{}/{}/data-{}.txt".format(mesure_location,number_pod, type_dns,type_code, delay,i)
        with open(filename) as f:
            contents = f.readlines()
            for data in contents:
                time_lookup , time_request , time_total = handle.extract_timing_info(data)
                list_time_lookup.append(time_lookup)
                list_time_request.append(time_request)
                list_time_total.append(time_total)
        f.close
    # calculate averange data
    time_lookup_mean = np.mean(list_time_lookup)
    time_request_mean = np.mean(list_time_request)
    time_total_mean = np.mean(list_time_total)

    time_lookup_error = np.std(list_time_lookup) / np.sqrt(len(list_time_lookup))
    time_request_error = np.std(list_time_request) / np.sqrt(len(list_time_request))
    time_total_error = np.std(list_time_total) / np.sqrt(len(list_time_total))
    return DataBlockByDelay(DataBlock(time_lookup_mean, time_lookup_error), 
                            DataBlock(time_request_mean, time_request_error), 
                            DataBlock(time_total_mean,time_total_error))    
```

```python
def calculate_mean_and_error_from_list(mesure_location , number_pod, type_dns, type_code, delay, pod_location=None):
    list_time_lookup = []
    list_time_request = []
    list_time_total = []
    
    for i in range(1,3):
        filename = ""
        if number_pod == 1:
            filename = "./{}/{}_{}_{}/{}/{}/data-{}.txt".format(mesure_location,number_pod, type_dns, pod_location,type_code, delay,i)
        else:
            filename = "./{}/{}_{}/{}/{}/data-{}.txt".format(mesure_location,number_pod, type_dns,type_code, delay,i)
        with open(filename) as f:
            contents = f.readlines()
            for data in contents:
                time_lookup , time_request , time_total = handle.extract_timing_info(data)
                list_time_lookup.append(time_lookup)
                list_time_request.append(time_request)
                list_time_total.append(time_total)
        f.close
    # calculate averange data
    time_lookup_mean = np.mean(list_time_lookup)
    time_request_mean = np.mean(list_time_request)
    time_total_mean = np.mean(list_time_total)

    time_lookup_error = np.std(list_time_lookup) / np.sqrt(len(list_time_lookup))
    time_request_error = np.std(list_time_request) / np.sqrt(len(list_time_request))
    time_total_error = np.std(list_time_total) / np.sqrt(len(list_time_total))
    return DataBlockByDelay(DataBlock(time_lookup_mean, time_lookup_error), 
                            DataBlock(time_request_mean, time_request_error), 
                            DataBlock(time_total_mean,time_total_error))    
```
